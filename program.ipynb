```python
# 1.Addiction of elements
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students.append("eve")
print(students)
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students.insert(4,"james")
print(students)
```

```python
#Remove element
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students.remove("ayushi")
print(students)
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students.pop(2)
print(students)
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
del students[5]
print(students)
```

```python
# changing of elements
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students[3]="rachel"
print(students)
```

```python
students=["ayushi","tina","ryuk","bob","dolly","slur","rem","john","kelly","lucy"]
students[2:5]=["olivia","sakshi"]
print(students)
```

```python
# 2. interchange the list items
```

```python
number=[10,20,30,40,50,60]
print("before swapping:",number)
temp=number[0]  
number[0]=number[-2]
number[-2]=temp
print("after swapping:",number)
```

```python
# 3. create a dict
```

```python
dict={"name":"ayushi","age":18,"course":"data science","city":"vrindavan"}
print(dict)
```

```python
# add element
```

```python
dict={"name":"ayushi","age":18,"city":"vrindavan"}
dict["age"]=19
dict["course"]="BSCDS"
print(dict)
```

```python
# remove element
```

```python
dict={"name":"ayushi","age":18,"course":"data science","city":"vrindavan"}
del dict["course"]
print(dict)
```

```python
# copy item
```

```python
dict={"name":"ayushi","age":18,"course":"data science","city":"vrindavan"}
dict.copy()
print(dict)
```

```python
# create nested dictionaries
```

```python
dict={
    "student1":{"name":"john","age":21,"city":"paris"},
    "student2":{"name":"lucy","age":20,"city":"chicago"}
}
print(dict)
                
```
